### 1) What is EDA ? What are steps in EDA Explain?

#### Solution

* EDA stands for Exploratory Data Analysis. It is the process of analyzing and visualizing data to gain insights, understand patterns, and summarize the main characteristics of the dataset. EDA helps in understanding the data, identifying patterns, detecting anomalies, and formulating hypotheses for further analysis. It is typically the first step in the data analysis process.

* The steps involved in EDA can vary depending on the specific dataset and goals of analysis. However, here are some common steps in the EDA process:

* Data Collection: Gather the relevant data from various sources, such as databases, spreadsheets, or APIs.

* Data Cleaning: Clean the data to ensure it is in a usable format. This may involve handling missing values, handling outliers, correcting data inconsistencies, and transforming data types.

* Data Exploration: Explore the data by examining its structure, dimensions, and properties. Calculate summary statistics, such as mean, median, standard deviation, and explore the distribution of variables.

* Univariate Analysis: Analyze individual variables one at a time to understand their distribution, range, and outliers. Visualize the data using histograms, box plots, bar plots, or other appropriate plots.

* Bivariate Analysis: Explore the relationship between two variables. This can involve creating scatter plots, correlation matrices, or other visualizations to identify any associations or patterns.

* Multivariate Analysis: Analyze the relationship between multiple variables simultaneously. This can include creating heatmaps, parallel coordinate plots, or other visualizations to understand complex interactions and correlations.

* Feature Engineering: Create new features or transform existing features to extract more meaningful information from the data. This may involve binning, scaling, one-hot encoding, or other techniques.

* Visualization: Use various visualizations, such as line plots, bar charts, scatter plots, or heatmaps, to represent the data and identify patterns or trends. Visualization helps in communicating insights and findings effectively.

* Statistical Modeling: Apply statistical techniques, such as regression, clustering, or hypothesis testing, to analyze the data and make inferences. This step goes beyond EDA and may involve more advanced analysis techniques.

* Documentation: Document the findings, insights, and limitations of the EDA process. This helps in communicating the results to stakeholders and provides a reference for future analysis.


### 2) What is outlayers? Explain the outlayers with formulas?

#### Solution :

* Outliers are data points that significantly deviate from the rest of the data in a dataset. They are observations that lie an abnormal distance away from other values. Outliers can occur due to various reasons such as measurement errors, experimental errors, or natural variations in the data.

* Outliers can have a significant impact on data analysis and statistical modeling because they can skew the results and distort the overall pattern. It is important to identify and handle outliers appropriately to ensure accurate and reliable analysis.

* IQR (Interquartile Range): The interquartile range is a measure of statistical dispersion, representing the range between the 25th and 75th percentiles of the data. Observations outside a certain range defined by the IQR are considered outliers. The formula for calculating the IQR is:

#### IQR = Q3 - Q1

* where Q1 is the 25th percentile (lower quartile) and Q3 is the 75th percentile (upper quartile).

* Outliers can be defined as values below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.

### 5) Problem Statement 2

Problem Statement:

The problem we aim to address is to understand the characteristics and patterns among different species of Iris flowers based on their measurements. By analyzing the Iris dataset, we want to explore the relationships between the features (sepal length, sepal width, petal length, and petal width) and the species of Iris flowers (setosa, versicolor, and virginica).

Specifically, we aim to:
1. Identify any distinct patterns or clusters among the Iris species based on the feature measurements.
2. Determine if there are significant differences in the distribution of features between the Iris species.
3. Explore the correlation between different features and investigate if certain features are strongly correlated with each other.
4. Analyze the relationship between the feature measurements and the species classification.

By conducting an EDA on the Iris dataset, we seek to gain insights that can contribute to a better understanding of the characteristics and relationships among different species of Iris flowers. These insights can potentially assist in species identification, classification, or further analysis in the field of botany or horticulture.

```python
# Solution

# Loading data set
import seaborn as sns
iris_df  = sns.load_dataset("iris")
iris_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Display the first few rows of the dataset
print(iris_df.head())

# Check the dimensions of the dataset
print(iris_df.shape)

# View the column names and data types
print(iris_df.info())
# Calculate summary statistics for numerical variables
print(iris_df.describe())

# Count the number of instances for each species
print(iris_df['species'].value_counts())
# Create a pair plot to visualize the relationships between variables
sns.pairplot(data=iris_df, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()

# Create box plots to compare the distribution of each feature by species
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
sns.boxplot(data=iris_df, x='species', y='sepal_length')
plt.subplot(2, 2, 2)
sns.boxplot(data=iris_df, x='species', y='sepal_width')
plt.subplot(2, 2, 3)
sns.boxplot(data=iris_df, x='species', y='petal_length')
plt.subplot(2, 2, 4)
sns.boxplot(data=iris_df, x='species', y='petal_width')
plt.suptitle('Distribution of Iris Features by Species')
plt.tight_layout()
plt.show()
# Compute the correlation matrix
correlation_matrix = iris_df.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Create violin plots to compare the distribution of each feature by species
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
sns.violinplot(data=iris_df, x='species', y='sepal_length')
plt.subplot(2, 2, 2)
sns.violinplot(data=iris_df, x='species', y='sepal_width')
plt.subplot(2, 2, 3)
sns.violinplot(data=iris_df, x='species', y='petal_length')
plt.subplot(2, 2, 4)
sns.violinplot(data=iris_df, x='species', y='petal_width')
plt.suptitle('Distribution of Iris Features by Species')
plt.tight_layout()
plt.show()
```
