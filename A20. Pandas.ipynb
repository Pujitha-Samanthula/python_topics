### What is pandas? 



```python
pandas---pannel datas
it is a open source library to work with relational data
it creates series and dataframes to work with datasets.
pandas supports different types of files like csv/xls and convert them into dataframes.
it is used to work with multidimensional data and data analysis and data manipulation.
pandas datastructures like series and dataframes are mutuable.
```

### Describe the different datastruchers in pandas?


```python
they are 2 types of datastructure
1.series
2.dataframe
1.series:
series is a one-dimensional labelled array capable of holding any type of data.
a column in excelsheet can also be referred as an series.
size is immutable and data is mutuable.
it can store homogeneous data only
2.dataframe:
dataframe is a two-dimensional labelled array capable of holding any type of data.
it is similar to spreadsheet/sql table/excel.
it contains columns and rows like a table.
size and data are mutuable.
it can store heterogeneous data.
```

### Explain the all steps in handleing outlayers?


```python
outliers:
the data point that doesnt belong to the group are called outliers.
steps:
1.sort the dataset in ascending order
2.find the Q1 and Q3 values
3.find the IQR=Q3-Q1
4.calculate the upperbound value(Q3+1.5*IQR)
5.calculate the lowerbound value(Q1-1.5*IQR)
6.check the dataset and identify the values who are far away from lowerbound and upperbound
7.these values are outliers(so we can drop those values or assign new values)
```
