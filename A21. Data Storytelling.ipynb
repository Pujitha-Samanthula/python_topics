### 4.Write a program to create a scotter plot on "total_bill" and "tip" columns using seaborn 

```python
import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill",y="tip",data=df,hue="total_bill",color=sns.color_palette())
plt.show()
```

### 5.Write a program to create a scotter plot on "total_bill" and "tip" columns using matplotlib with blue color datapoints

```python
import matplotlib.pyplot as plt
x=df.total_bill
y=df.tip
plt.scatter(x,y,label="tip")
plt.legend()
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.show()
```

### 6.Write a program to create a barplot on "day" and "total_bill" columns using seaborn

### 7.Write a program to create a barplot on "day" and "total_bill" columns using matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
x=df.day
y=df.total_bill
plt.bar(x,y,color="b",label="total_bill")
plt.xlabel="day"
plt.ylabel="total_bill"
plt.legend()
plt.show()
```

### 14.Write a program to calculate the sum of "tips" for each "time" period use groupby and create a pie chart using matplotlib

```python
import matplotlib.pyplot as plt
a=df.groupby("time")["tip"].sum()
print(a)
plt.pie(a,autopct='%.1f%%',shadow=True,labels=["lunch","dinner"],explode=[0.1,0])
plt.show()
```

### 15.Write a program to group the data by "day" and calculate the average "tip" and visualize using line plot with matplotlib

```python
import matplotlib.pyplot as plt
a=df.groupby(["day"])["tip"].mean()
print(a)
plt.plot(a,color="hotpink")
plt.xlabel("day")
plt.ylabel("tip")
```

### 16.Write a program to group the data by "day" and calculate the average "tip" and visualize using line plot with seaborn

```python
import matplotlib.pyplot as plt
a=df.groupby(["day"])["tip"].mean()
print(a)
x=df.day.unique()
print(x)
sns.lineplot(a,color="b")
plt.show()
```
